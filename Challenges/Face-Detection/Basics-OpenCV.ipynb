```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("google-logo.jpg")

newImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #Important
plt.figure()
plt.imshow(img)
plt.show()
plt.imshow(newImg)
plt.show()
# by default cv2 is based on Blue, green and red. hence color got swapped
# cv2.imshow("gray", gray)
```

## Image Dispay in different Window

```python
import cv2
img = cv2.imread("google-logo.jpg") # used for reading images
cv2.imshow("g-logo", img) # used for displaying image

gray = cv2.imread("google-logo.jpg", cv2.IMREAD_GRAYSCALE)
cv2.imshow("gray-logo", gray)

cv2.waitKey(0) # amount of time take to close it automatically and 0 will held until you close
cv2.destroyAllWindows()
```

- its plot will be

```python
import matplotlib.pyplot as plt
plt.imshow(img)
plt.show()
plt.imshow(gray)
plt.show()
```
